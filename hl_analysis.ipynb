```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('hl_results.csv')
```

```python
results.head()
```

```python
results['net_return'].hist(bins='auto')
plt.xlabel('Net Return')
plt.ylabel('Frequency')
plt.title('Histogram of Net Return')
plt.show()
```

```python
results['avg_return_per_hand'].hist(bins='auto')
plt.xlabel('Avg Return Per Hand')
plt.ylabel('Frequency')
plt.title('Histogram of Avg Return')
plt.show()
```

```python
net_mean = results['net_return'].mean()
net_median = results['net_return'].median()
avg_mean = results['avg_return_per_hand'].mean()
avg_median = results['avg_return_per_hand'].median()
```

```python
print(f"Net Return Results --- Mean: {net_mean}, Median: {net_median}")
print(f"Avg Return Results --- Mean: {avg_mean}, Median: {avg_median}")
```
